#### Introduction to Streamlit
Streamlit is an open-source app framework for Machine Learning and Data Science projects. It allows you to create beautiful web applications for your machine learning and data science projects with simple Python scripts.

1. Create  a "requirments.txt" file and list all of the necessary dependencies.
2. To download dependencies run -> pip install -r requirements.txt

Key concepts (interview-ready)

Script-as-app: Streamlit re-executes the script from top to bottom on each user interaction; widgets return values.

Widgets: st.slider, st.selectbox, st.button, st.file_uploader, st.chat_input, etc.

Caching:

@st.cache_data for pure data functions (e.g., pd.read_csv, expensive queries).

@st.cache_resource for heavy objects (models, DB connections).

State: st.session_state persists values across reruns per user session.

Layout: columns st.columns, tabs st.tabs, sidebar st.sidebar, wide mode via st.set_page_config.

Output: tables (st.dataframe), charts (st.line_chart, st.bar_chart, or use Altair/Plotly/Matplotlib figures).

Deployment: Streamlit Community Cloud, Docker+any cloud (EC2, Cloud Run), or behind an internal reverse proxy.
